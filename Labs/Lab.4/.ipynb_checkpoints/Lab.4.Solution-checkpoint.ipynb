1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class counter:
    def __init__(self, max_value = 5):
        self.__current_value = 0
        self.__max_value = max_value

    #getters
    def get_current_value(self):
        return self.__current_value
    def get_max_value(self):
        return self.__max_value

    #methods
    def increment_value(self, n):
        if isinstance(n,(int,float)):
            self.__current_value +=n
            if self.__current_value >= self.__max_value:
                self.__current_value = self.__max_value
                print('max value has been reached. ur input is', n - self.__max_value, 'over the max')
        else:
            print('u must enter an integer')
    def reset_value(self, n = 0):
        self.__current_value = n
        print('your counter has been reset')
    def print_counter(self):
        print(self.__current_value)
        
```

```python
counter_1 = counter()
counter_1.increment_value(2)
```

```python
counter_1.print_counter()
```

```python
class rectangle:
    def __init__(self,length,width,x,y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
    #getters
    def get_length(self):
        return self.__length
    def get_width(self):
        return self.__width
    def get_x(self):
        return self.__x
    def get_y(self):
        return self.__y

    #methods
    def compute_perimeter(self):
        return (2*(self.__length)) + (2*(self.__width))
    def compute_area(self):
        return self.__length * self.__width
```

```python
#copied canvas class from lecture
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class Shape:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def compute_area(self):
        raise NotImplementedError
    
    def compute_perimeter(self):
        raise NotImplementedError
    
    def get_perimeter_points(self, num_points=16):
        raise NotImplementedError

    def is_inside_object(self, x, y):
        raise NotImplementedError

    def overlaps_with(self, other):
        for point in other.get_perimeter_points():
            if self.is_inside_object(*point):
                return True
        for point in self.get_perimeter_points():
            if other.is_inside_object(*point):
                return True
        return False

    def paint(self, canvas, char='*'):
        for point in self.get_perimeter_points(100):
            canvas.set_pixel(int(point[1]), int(point[0]), char)

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        super().__init__(x, y)
        self.__length = length
        self.__width = width
    
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def compute_perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def compute_area(self):
        return self.__length * self.__width
    
    def get_perimeter_points(self, num_points=16):
        points = []
        sides = [(0, 0, self.__length, 0), 
                 (self.__length, 0, self.__length, self.__width),
                 (self.__length, self.__width, 0, self.__width),
                 (0, self.__width, 0, 0)]
        points_per_side = max(1, num_points // 4)
        
        for start_x, start_y, end_x, end_y in sides:
            for i in range(points_per_side):
                t = i / (points_per_side - 1) if points_per_side > 1 else 0
                x = start_x + t * (end_x - start_x) + self.get_x()
                y = start_y + t * (end_y - start_y) + self.get_y()
                points.append((x, y))
        
        return points[:num_points]

    def is_inside_object(self, x, y):
        return (self.get_x() <= x <= self.get_x() + self.__length and
                self.get_y() <= y <= self.get_y() + self.__width)

class Circle(Shape):
    def __init__(self, radius, x, y):
        super().__init__(x, y)
        self.__radius = radius
    
    def get_radius(self):
        return self.__radius
    
    def compute_area(self):
        return math.pi * self.__radius ** 2
    
    def compute_perimeter(self):
        return 2 * math.pi * self.__radius
    
    def get_perimeter_points(self, num_points=16):
        points = []
        for i in range(num_points):
            angle = 2 * math.pi * i / num_points
            x = self.get_x() + self.__radius * math.cos(angle)
            y = self.get_y() + self.__radius * math.sin(angle)
            points.append((x, y))
        return points

    def is_inside_object(self, x, y):
        distance = math.sqrt((x - self.get_x())**2 + (y - self.get_y())**2)
        return distance <= self.__radius

class Triangle(Shape):
    def __init__(self, base_s1, s2, s3, height, x, y):
        super().__init__(x, y)
        self.__base_s1 = base_s1
        self.__height = height
        self.__s2 = s2
        self.__s3 = s3
    
    def get_base_s1(self):
        return self.__base_s1
    
    def get_height(self):
        return self.__height
    
    def get_s2(self):
        return self.__s2
    
    def get_s3(self):
        return self.__s3
    
    def compute_area(self):
        return 0.5 * self.__base_s1 * self.__height
    
    def compute_perimeter(self):
        return self.__base_s1 + self.__s2 + self.__s3
    
    def get_perimeter_points(self, num_points=16):
        points = []
        sides = [self.__base_s1, self.__s2, self.__s3]
        total_perimeter = sum(sides)
        
        current_distance = 0
        for side_length in sides:
            side_points = max(1, int(num_points * side_length / total_perimeter))
            for i in range(side_points):
                t = i / side_points
                angle = 2 * math.pi * (current_distance + t * side_length) / total_perimeter
                x = self.get_x() + (total_perimeter / (2 * math.pi)) * math.cos(angle)
                y = self.get_y() + (total_perimeter / (2 * math.pi)) * math.sin(angle)
                points.append((x, y))
            current_distance += side_length
        return points[:num_points]

    def is_inside_object(self, x, y):
        # Simplified implementation, assuming the triangle is equilateral
        # For a more accurate implementation, you'd need to use barycentric coordinates
        return (x >= self.get_x() and x <= self.get_x() + self.__base_s1 and
                y >= self.get_y() and y <= self.get_y() + self.__height)

class CompoundShape(Shape):
    def __init__(self, shapes):
        super().__init__(min(shape.get_x() for shape in shapes),
                         min(shape.get_y() for shape in shapes))
        self.shapes = shapes

    def compute_area(self):
        return sum(shape.compute_area() for shape in self.shapes)

    def compute_perimeter(self):
        return sum(shape.compute_perimeter() for shape in self.shapes)

    def get_perimeter_points(self, num_points=16):
        all_points = []
        for shape in self.shapes:
            all_points.extend(shape.get_perimeter_points(num_points // len(self.shapes)))
        return all_points[:num_points]

    def is_inside_object(self, x, y):
        return any(shape.is_inside_object(x, y) for shape in self.shapes)

    def paint(self, canvas, char='*'):
        for shape in self.shapes:
            shape.paint(canvas, char)

def create_simple_drawing():
    canvas = Canvas(40, 20)
    
    rectangle = Rectangle(10, 5, 5, 5)
    circle = Circle(3, 20, 10)
    triangle = Triangle(6, 6, 6, 4, 30, 15)
    
    compound_shape = CompoundShape([Rectangle(3, 3, 25, 2), Circle(2, 28, 5)])
    
    rectangle.paint(canvas, 'R')
    circle.paint(canvas, 'C')
    triangle.paint(canvas, 'T')
    compound_shape.paint(canvas, 'S')
    
    canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def paint(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.paint(self.canvas)
        self.canvas.display()

    def modify_shape(self, old_shape, new_shape):
        if old_shape in self.shapes:
            index = self.shapes.index(old_shape)
            self.shapes[index] = new_shape

def demonstrate_raster_drawing():
    # Create a RasterDrawing
    drawing = RasterDrawing(40, 20)

    
    rectangle = Rectangle(10, 5, 5, 5)
    circle = Circle(3, 20, 10)
    triangle = Triangle(6, 6, 6, 4, 30, 15)
    compound_shape = CompoundShape([Rectangle(3, 3, 25, 2), Circle(2, 28, 5)])

    drawing.add_shape(rectangle)
    drawing.add_shape(circle)
    drawing.add_shape(triangle)
    drawing.add_shape(compound_shape)

    print("Initial drawing:")
    drawing.paint()

    print("\nModifying the drawing...")
    drawing.remove_shape(triangle)
    drawing.modify_shape(rectangle, Rectangle(8, 4, 7, 7))
    new_circle = Circle(4, 18, 8)
    drawing.add_shape(new_circle)

    print("Modified drawing:")
    drawing.paint()


demonstrate_raster_drawing()
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
